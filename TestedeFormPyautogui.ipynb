```python
# O objetivo deste código é automatizar o processo de preenchimento de formulário.
# Usaremos além do pacote PyAutoGui, A Library Pandas para manipulação de dados com Python.
# Leremos o arquivo do excel para que ele seja realocado no formulário web.


#Importação dos dados

import pyautogui, os
import pandas as pd
from time import sleep


"""
O primeiro passo é ler o arquivo Excel 
Configurar o botão enviar
Configurar os dados do excel para o formulário
"""

# Realizar função
caminho = r'C:\Users\luizg\OneDrive\Documentos\PROJETOPYAUTOGUI'
os.chdir(caminho)
botao = 'botao.png'
arquivo = 'base.xlsx'
dados = pd.read_excel(arquivo)

cmp_nome = (590, 475)
cmp_escolha1 = (592, 653)
cmp_escolha2 = (590, 701)
cmp_escolha3 = (591, 757)
cmp_texto1 = (611, 459)
cmp_aval1 = (728, 693)
cmp_aval3 = (948, 693)
cmp_aval5 = (1164, 691)
cmp_resposta = (661, 324)

# Preencher os campos necessários
# A função range é responsável por retornar uma sequência de números começando por 0 como padrão.
# até o número ou tamanho desejado, no nosso caso, o tamanho "len" do campo selecionado.
# 


for i in range (0, len(dados.NOME)):
    sleep(1)

    """
    Move a seta do mouse e clica no campo de nome
    """

    pyautogui.moveTo(cmp_nome, duration=0.5)
    pyautogui.click()
    sleep(1)
    
    """
    essa função vai pegar o nome e opções de escolha do excel e digitar no campo clicado. 
    """
    pyautogui.typewrite(dados.iloc[i]['NOME'], interval=0.10)
    sleep(1)

    if dados.iloc[i]['ESCOLHA'] == 1:
        pyautogui.moveTo(cmp_escolha1, duration=0.5)
    elif dados.iloc[i]['ESCOLHA'] == 2:
        pyautogui.moveTo(cmp_escolha2, duration=0.10)
    else:
        pyautogui.moveTo(cmp_escolha3, duration=0.10)   
    pyautogui.click()
    sleep(1)

    """
    Scroll do mouse para completar o resto do formulário
    """
    pyautogui.scroll(-600)
    sleep(2)

    """ 
    bloco de texto para digitação
    """
    pyautogui.moveTo(cmp_texto1, duration=0.5)
    pyautogui.click()
    sleep(1)
    pyautogui.typewrite(dados.iloc[i]['TEXTO'], interval=0.10)
    sleep(1)

    """
    Avaliação 
    """
    if dados.iloc[i]['AVALIAÇÃO'] == 5:
        pyautogui.moveTo(cmp_aval1, duration=0.5)
    elif dados.iloc[i]['AVALIAÇÃO'] == 4:
        pyautogui.moveTo(cmp_aval3, duration=0.5)
    else:
        pyautogui.moveTo(cmp_aval5, duration=0.5)

    pyautogui.click()
    sleep(0.5)

    botaoclicado = pyautogui.locateCenterOnScreen(botao)
    pyautogui.moveTo(botaoclicado, duration = 0.5)
    pyautogui.click()
    sleep(2)

    pyautogui.moveTo(cmp_resposta, duration=0.5)
    pyautogui.click()
    sleep(2)


```
